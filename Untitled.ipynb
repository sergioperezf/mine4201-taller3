```python
import pandas as pd
import pickle
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
f = open('movie_data.pkl', 'rb')
movie_data = pickle.load(f)
f.close()
```

```python
len(movie_data)
```

```python
f = open('ratings.pkl', 'rb')
ratings = pickle.load(f)
f.close()
```

```python
print(len(ratings))
ratings_filtered = []
```

```python

for rat in ratings:
    if int(rat[1]) in movie_data:
#         print(type(rat))
        ratings_filtered.append(rat)
    
```

```python
len(ratings_filtered)
```

```python
rf = np.array(ratings_filtered)
```

```python
len(rf)
```

```python
movie_data_trimmed = {}
for rating in rf:
    i = int(rating[1])
    if i in movie_data:
        movie_data_trimmed[i] = movie_data[i]
```

```python
len(movie_data_trimmed)
```

```python
pickle.dump(rf, open('ratings.pkl', 'wb'))
```

```python
data = {'id': [], 'text': []}
for mov_id in movie_data_trimmed.keys():
    data['id'].append(mov_id)
    data['text'].append(movie_data_trimmed[mov_id]['data'][0]['abstract']['value'])
```

```python
df = pd.DataFrame(data=data)
```

```python
tf = TfidfVectorizer(analyzer='word', min_df=0.1, stop_words='english')
```

```python
tfidf_matrix = tf.fit_transform(df['text'])
```

```python
tfidf_matrix[df.loc[df['id'] == 31].index[0]].toarray()[0]
```

```python
tf_idf_arrays = {}
for movie_id in movie_data_trimmed.keys():
    tf_idf_arrays[movie_id] = tfidf_matrix[df.loc[df['id'] == movie_id].index[0]].toarray()[0]
```

```python
import pickle
```

```python
pickle.dump(tf_idf_arrays, open('movies_tfidf.pkl', 'wb'))
```
